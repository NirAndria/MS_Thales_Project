```python
lemmatizer = WordNetLemmatizer()
from nltk.corpus import stopwords

stop_words = set((" ".join(stopwords.words('english'))).lower().split(' '))


def lemmatizer_fct(text):

  return " ".join(list(map(lambda word: lemmatizer.lemmatize(word.lower(), pos="v"), text.split()))).translate(str.maketrans('', '', string.punctuation))

def rm_stop_word(text):
  set_input = set(text.split(" "))

  clean_set = set_input.difference(stop_words)

  return " ".join(list(clean_set))


def global_cleaning(text):
  return rm_stop_word(lemmatizer_fct(text))

text_reference = global_cleaning(text_reference)
text_generated = global_cleaning(text_generated)

```

```python
# Initialize the TfidfVectorizer
vectorizer = TfidfVectorizer()

# Combine the reference and validation texts into a list
texts = [text_reference, text_generated]

# Fit and transform the texts to obtain TF-IDF vectors
tfidf_matrix = vectorizer.fit_transform(texts)

# Compute cosine similarity between the reference and validation text
cos_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])

# Output the TF-IDF matrix
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())

# Output the Cosine Similarity
print(f"\nCosine Similarity between reference and validation text: {cos_sim[0][0]}")

# To view the TF-IDF scores for each word, we can get the terms
terms = vectorizer.get_feature_names_out()

# Print each term and its TF-IDF score for the reference and validation text
print("\nTF-IDF scores for each word:")
for i, term in enumerate(terms):
    print(f"{term}: Reference Text: {tfidf_matrix[0, i]:.4f}, Validation Text: {tfidf_matrix[1, i]:.4f}")
```

```python
def get_bert_embedding(sentence):
    # Load pre-trained BERT model and tokenizer
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    model = BertModel.from_pretrained('bert-base-uncased')

    # Tokenize input sentence
    inputs = tokenizer(sentence, return_tensors='pt', padding=True, truncation=True, max_length=512)
    #print(inputs)
    # Get embeddings from BERT
    with torch.no_grad():
        outputs = model(**inputs)
        #print(outputs)

    # Mean pooling to get a sentence-level embedding
    sentence_embedding = outputs.last_hidden_state.mean(dim=1).squeeze()
    #print(sentence_embedding)
    return sentence_embedding

def compare_strings(string1, string2):
    # Generate embeddings for both strings
    embedding1 = get_bert_embedding(string1).unsqueeze(0)  # Add batch dimension
    embedding2 = get_bert_embedding(string2).unsqueeze(0)

    # Compute cosine similarity
    similarity = cosine_similarity(embedding1.numpy(), embedding2.numpy())[0][0]

    return similarity





text_reference1 = "Tugce Tunalilar'09 on the Stage with Pink Martini tarafindan Arzum Kopsa | May 30, 2018 | News | 0 yorum Our alumni Tugce Tunalilar was on the stage with Pink Martini on May 22 at Princeton University. Tugce works at Crossix in New York. Her interest in music started when she was a student at Hisar thanks to Head of Music Department Levent Atlier's guidance and effort. Our alumni says the business life in New York City is busy and, she likes making music. She has also sang with Demir Demirkan during his visit to the U.S. and helped him to contact the Turkish Student groups there. She worked with a few music groups in Brooklyn and New York and she is currently planning to record the Greek Eurivision song 'Oneiro Mou'. We congratulate her wish success in her efforts. Click for the videos."
text_generated1 ="Tugce Tunalilar'09 on the Stage with Pink Martini tarafindan Arzum Kopsa | May 30, 2018 | News | 0 yorum Our alumni Tugce Tunalilar was on the stage with Pink Martini on May 22 at Princeton University. Tugce works at Crossix in New York. Her interest in music started when she was a student at Hisar thanks to Head of Music Department Levent Atlier's guidance and effort. Our alumni says the business life in New York City is busy and, she likes making music. She has also sang with Demir Demirkan during his visit to the U.S. and helped him to contact the Turkish Student groups there. She worked with a few music groups in Brooklyn and New York and she is currently planning to record the Greek Eurivision song 'Oneiro Mou''. We congratulate her wish success in her efforts. Click for the videos."
similarity_score = compare_strings(text_reference1,text_generated1)
print(f"Similarity Score: {similarity_score:.4f}")
```
